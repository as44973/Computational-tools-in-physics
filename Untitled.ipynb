```python
import numpy as np
import scipy.integrate as scii
import matplotlib.pyplot as plt

def posevni_met(t, y, upor):
    v_ksi = y[2]
    v_zeta = y[3]
    abs_v = np.sqrt(v_ksi**2 + v_zeta**2)
    return np.array([v_ksi, v_zeta, -upor*abs_v*v_ksi, -1 -upor*abs_v*v_zeta])

A = 2
fi = np.pi / 4
start = 0.0
stop = 10
ts = np.linspace(start, stop, 1000)
y_init = np.array([0.0, 0.0, A*np.cos(fi), A*np.sin(fi)])

def solve_ivp():
    def position_zero(t, y, *args):
        return y[1]
    position_zero.terminal=True
    position_zero.direction=-1.0

    ksi = []
    zeta = []
    for upor in [0, 1]:
        solve = scii.solve_ivp(posevni_met, (start, stop), y_init, method='RK45', t_eval=ts, events=position_zero, args=(upor,))
        ksi.append(solve.y[0])
        zeta.append(solve.y[1])

    plt.plot(ksi[0], zeta[0])
    plt.plot(ksi[1], zeta[1])
    plt.show()
    plt.close()
solve_ivp()

def solve_ivp_domet():
    def position_zero(t, y, args):
        return y[1]
    position_zero.terminal=True
    position_zero.direction=-1.0

    A = 4*np.linspace(0.01, 0.99, 100)
    domet = []
    domet_upor = []
    for Ai in A:
        y_init = np.array([0.0, 0.0, Ai*np.cos(fi), Ai*np.sin(fi)])
        for upor in [0, 1]:
            solve_domet = scii.solve_ivp(posevni_met, (start, stop), y_init, method='RK23', t_eval=ts, events=position_zero, args=(upor,))
            if upor == 0:
                domet.append(solve_domet.y[0][-1])
            if upor == 1:
                domet_upor.append(solve_domet.y[0][-1])
    plt.plot(A, domet)
    plt.plot(A, domet_upor)
    plt.legend()
    plt.show()
solve_ivp_domet()
```
