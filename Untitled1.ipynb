```python
#!/usr/bin/env python3

import numpy as np
import scipy.integrate as scii
import matplotlib.pyplot as plt

def posevni_met_ksi(t, y, A, fi):
    return np.array([y[1], -A*np.cos(fi)*y[1]])

def posevni_met_zeta(t, y, A, fi):
    return np.array([y[1], -y[0]-A*np.sin(fi)*y[1]])


start = 0.0
stop = 100.0
y_init = np.array([0.0, np.sqrt(2)])

def solve_ivp():
    A = 2
    fi = np.pi / 4
    solve_ksi = scii.solve_ivp(posevni_met_ksi, (start, stop), y_init, method='RK23', args=(A, fi), t_eval=np.linspace(0, 100, 100000))
    solve_zeta = scii.solve_ivp(posevni_met_zeta, (start, stop), y_init, method='RK23', args=(A, fi), t_eval=np.linspace(0, 100, 100000))
    
    ksi = solve_ksi.y[0]
    zeta = solve_zeta.y[0]
    
    plt.plot(ksi, zeta, label='Bruh')
    plt.show()
solve_ivp()
```
