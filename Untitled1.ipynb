```python
#ANANDRAZ '''#dreamteam'''
import numpy as np
import random

identiteta = [0, 0, 0, 0, 0, 0, 0, 0, 0]
# id.number, Dovzeten, Kužen, Bolan, Odporen, identiteta celice, Tip celice, št. dni okužbe, št. dni prebolevanja,

p = 22550 #populacija

st_ljudi = [identiteta] * p
novo_st_ljudi = []

for i in range(1, p + 1):
    novo_st_ljudi.append([i, 1, 0, 0, 0, 0, 0, 0, 0])

#Celice
N_1 = 2700
N_2 = 2000
N_3 = 1500
N_4 = 1200
N_5 = 400
N_6 = 300
N_7 = 200
N_8 = 100
N_9 = 50
N_10 = 10  

N = [N_1, N_2, N_3, N_4, N_5, N_6, N_7, N_8, N_9, N_10]        
C = []        
for i in range(1,11):
    c = []
    l = 1
    for j in range(N[i-1] * i):
        j = j + 1
        if (j % i) == 0:
            c.append([0, 0, 0, 0, 0, l, i, 0, 0])
            l += 1
        else:
            c.append([0, 0, 0, 0, 0, l, i, 0, 0])
    C += c

novo_st_ljudi = np.array(novo_st_ljudi) + np.array(C)

print(novo_st_ljudi)

alfa_1 = 0.45
alfa_2 = 0.10


#verjetnost obolevanja po dnevih:
beta1 = 0.01
beta2 = 0.03
beta3 = 0.05
beta4 = 0.08
beta5 = 0.12
beta6 = 0.10
beta7 = 0.07
beta8 = 0.04
beta9 = 0.02
beta10 = 0.01

sez_bet = [beta1, beta2, beta3, beta4, beta5, beta6, beta7, beta8, beta9, beta10]

dovzetni = []
kuzni = []
bolni = []
odporni = []

st_dni = 50
st_kuznih = 2 #kukr jih hocs na zacetku met okuzenih
st_dovzetnih = p - st_kuznih
st_bolnih = 0
st_odpornih = 0
indeksi_okuz = [] 


a = np.array([])
b = np.array([])
volare = np.array([])
d = np.array([])
e = np.array([])

se_ne_okuzeni = np.arange(1, p + 1)
se_ne_okuzeni = se_ne_okuzeni.tolist()


for i in range(st_kuznih):
    choice_indices = np.random.choice(se_ne_okuzeni)
    se_ne_okuzeni.remove(choice_indices)
    novo_st_ljudi[choice_indices - 1][1] = 0
    novo_st_ljudi[choice_indices - 1][2] = 1
    indeksi_okuz.append((novo_st_ljudi[choice_indices - 1][5], novo_st_ljudi[choice_indices - 1][6]))
   

   
for j in range(1, st_dni+1): #st dni pandemije
#ena velka for zanka za st dni. Mogoce bi se dalo z while zanko se vsakic en dan dodat?
    print(str(j+1) + " DAN")

   
    novi_kuzni = round((alfa_1 * st_kuznih + alfa_2 * st_bolnih) * st_dovzetnih/p) #na koncu zanke jih moras se razvrstit
    #izbere nakljucne ljudi ki jih okuzimo na podlagi st iz prejsnjega dneva in jih uvrstimo med kuzne
    #poiscemo id celic novo okuzenih in da okuzimo vse clane celic

   
    for i in range(novi_kuzni):
        if len(se_ne_okuzeni) > 0:
            choice_indices = np.random.choice(se_ne_okuzeni)
            se_ne_okuzeni.remove(choice_indices)
            novo_st_ljudi[choice_indices - 1][1] = 0
            novo_st_ljudi[choice_indices - 1][2] = 1
            for k in range(len(novo_st_ljudi)):
                if (novo_st_ljudi[k][5], novo_st_ljudi[k][6]) == (novo_st_ljudi[choice_indices - 1][5], novo_st_ljudi[choice_indices - 1][6]):
                    indeksi_okuz.append((novo_st_ljudi[choice_indices - 1][5], novo_st_ljudi[choice_indices - 1][6]))
                    novo_st_ljudi[k][1] = 0
                    novo_st_ljudi[k][2] = 1  # kako narest, da se ne bodo ponavljali?
                    kuzni.append(novo_st_ljudi[k])
                   
                    if novo_st_ljudi[k][0] in se_ne_okuzeni:
                        se_ne_okuzeni.remove(novo_st_ljudi[k][0])
                        st_dovzetnih -= 1
                        st_kuznih += 1
            st_dovzetnih -= 1
            st_kuznih += 1
           

                                       

   
    #BOLNI
    #okuzene razvrstimo po st dni inkubacije v seznam, ki ga pomnozimo s seznamom bet
   
    sez_kuznih = [0] * 10
    #dodajanje dni kuznim in bolnim na koncu zanke/dneva
    for c in range(len(novo_st_ljudi)):
        #inkubacijska doba za kuzne
        if novo_st_ljudi[c][2] == 1:
            novo_st_ljudi[c][7] += 1
            sez_kuznih[novo_st_ljudi[c][7] - 1] += 1
        #ce po 10 dneh ne zboli postane odporen
        if novo_st_ljudi[c][7] == 10:
            novo_st_ljudi[c][7] = 0
            novo_st_ljudi[c][2] = 0
            novo_st_ljudi[c][4] = 1
            odporni.append(novo_st_ljudi[c])
            st_kuznih -= 1
            st_odpornih += 1
        #inkubacijska doba za bolne
        if novo_st_ljudi[c][3] == 1:
            novo_st_ljudi[c][8] += 1
        #po 15 dneh gre v grupo odpornih
        if novo_st_ljudi[c][8] == 15:
            novo_st_ljudi[c][8] = 0
            novo_st_ljudi[c][4] = 1
            novo_st_ljudi[c][3] = 0
            odporni.append(novo_st_ljudi[c][0])
            st_odpornih += 1
            st_bolnih -= 1

    for i in range(len(sez_kuznih)):
        novo_bolni = round(sez_kuznih[i] * sez_bet[i])
        sez_moznih = []
        for g in range(len(novo_st_ljudi)):
            if (novo_st_ljudi[g][7] - 1) == i and (novo_st_ljudi[g][4] != 1):
                sez_moznih.append(g)
        for h in range(novo_bolni):
            if len(sez_moznih) != 0:
                indeks = np.random.choice(sez_moznih)
                novo_st_ljudi[indeks][7] = 0
                novo_st_ljudi[indeks][8] = 1
                novo_st_ljudi[indeks][2] = 0
                novo_st_ljudi[indeks][3] = 1
                bolni.append(indeks)
                sez_moznih.remove(indeks)
                st_kuznih -= 1
                st_bolnih += 1
        
    a = np.append(a, [st_dovzetnih])
    b = np.append(b, [st_kuznih])
    volare = np.append(volare, [st_bolnih])
    d = np.append(d, [st_odpornih])
    e = np.append(e, [j])
    

    
import matplotlib.pyplot as plt
plt.ylabel('st. dovzetnih')
plt.xlabel('st. dni')
plt.plot(e, a, 'o', label = 'Dovzetni')
plt.legend()
plt.savefig('dovzetni_1')
plt.show()

plt.ylabel('st. kuznih')
plt.xlabel('st. dni')
plt.plot(e, b, 'o', label = 'Kuzni')
plt.legend()
plt.savefig('kuzni_1')
plt.show()

plt.ylabel('st. bolnih')
plt.xlabel('st. dni')
plt.plot(e, volare, 'o', label = 'Bolni')
plt.legend()
plt.savefig('bolni_1')
plt.show()

plt.ylabel('st. odpornih')
plt.xlabel('st. dni')
plt.plot(e, d, 'o', label = 'Odporni')
plt.legend()
plt.savefig('odporni_1')
plt.show()
```
